```python
import pandas as pd
import requests
import json
breeze_key = '5464e85fcd80467aa68843e089c3a591'
chucks_key = 'ed6db1443cb54cc8af83958977a03b49'
lesly_key = 'e32c954207d84b24b198fb9cd883898e'
vijay_key = 'e2df0b746770457d8b860ec0ff24eb79'
new_key ='0018cfa0ded1415581537c9513992c54'
```

```python
city_info = pd.read_csv('coordinate_data.csv')
```

```python
#empty list to hold the info we want from the breezometer API
city = []
missing_cities = []
aqi = []
dominant_pol = []
category = []
#pollutants
pol_co = []
pol_o3 = []
pol_pm10 = []
pol_pm25 = []
pol_so2 = []  
pol_no2 = []
#pollutant units 
# as per US_EPA AOI standards, concentrations of gases are measured in ppb (parts per billion), particulate matter is measured in ug/m^3 
time = []

#run the first for loop using Alex's Key
for i in range(700):
    url = f'https://api.breezometer.com/air-quality/v2/current-conditions?lat={city_info.iloc[i,2]}&lon={city_info.iloc[i,3]}&key={lesly_key}&features=breezometer_aqi,dominant_pollutant_concentrations,sources_and_effects,pollutants_concentrations,pollutants_aqi_information'
    response = requests.get(url).json()
    if response['error'] == None:
        print(f'Processing record #{city_info.iloc[i,0]} for {city_info.iloc[i,1]}')
        try:
            city.append(city_info.iloc[i,1])
            aqi.append(response['data']['indexes']['baqi']['aqi'])
            dominant_pol.append(response['data']['indexes']['baqi']['dominant_pollutant'])
            category.append(response['data']['indexes']['baqi']['category'])
            pol_co.append(response['data']['pollutants']['co']['concentration']['value'])
            pol_o3.append(response['data']['pollutants']['o3']['concentration']['value'])
            pol_pm10.append(response['data']['pollutants']['pm10']['concentration']['value'])
            pol_pm25.append(response['data']['pollutants']['pm25']['concentration']['value'])
            pol_so2.append(response['data']['pollutants']['so2']['concentration']['value'])  
            pol_no2.append(response['data']['pollutants']['no2']['concentration']['value'])
            time.append(response['data']['datetime'])
        except: 
            print(f'Something went wrong getting the data for #{city_info.iloc[i,0]}, {city_info.iloc[i,1]}')
    else:
        print(f'No record for #{city_info.iloc[i,0]}, {city_info.iloc[i,1]}')
        missing_cities.append(city_info.iloc[i,1])
    
    
        
        
```

```python
#run second loop using chucks_key 
for i in range(700,1490):
    url = f'https://api.breezometer.com/air-quality/v2/current-conditions?lat={city_info.iloc[i,2]}&lon={city_info.iloc[i,3]}&key={vijay_key}&features=breezometer_aqi,dominant_pollutant_concentrations,sources_and_effects,pollutants_concentrations,pollutants_aqi_information'
    response = requests.get(url).json()
    if response['error'] == None:
        print(f'Processing record #{city_info.iloc[i,0]} for {city_info.iloc[i,1]}')
        try:
            city.append(city_info.iloc[i,1])
            aqi.append(response['data']['indexes']['baqi']['aqi'])
            dominant_pol.append(response['data']['indexes']['baqi']['dominant_pollutant'])
            category.append(response['data']['indexes']['baqi']['category'])
            pol_co.append(response['data']['pollutants']['co']['concentration']['value'])
            pol_o3.append(response['data']['pollutants']['o3']['concentration']['value'])
            pol_pm10.append(response['data']['pollutants']['pm10']['concentration']['value'])
            pol_pm25.append(response['data']['pollutants']['pm25']['concentration']['value'])
            pol_so2.append(response['data']['pollutants']['so2']['concentration']['value'])  
            pol_no2.append(response['data']['pollutants']['no2']['concentration']['value'])
            time.append(response['data']['datetime'])
        except: 
            print(f'Something went wrong getting the data for #{city_info.iloc[i,0]}, {city_info.iloc[i,1]}')
    else:
        print(f'No record for #{city_info.iloc[i,0]}, {city_info.iloc[i,1]}')
        missing_cities.append(city_info.iloc[i,1])
```

```python
#create dictionary holding all our data lists
air_quality_dict ={
    'city': city,
    'aqi': aqi,
    'dominant_pol': dominant_pol,
    'category': category,
    'pol_co': pol_co,
    'pol_o3': pol_o3,
    'pol_pm10': pol_pm10,
    'pol_pm25': pol_pm25,
    'pol_so2': pol_so2,
    'pol_no2': pol_no2,
    'time': time
}

#convert the dictionary into a pandas dataframe called 'ca_air_quality_data'
ca_air_quality_data = pd.DataFrame(air_quality_dict)
```

```python
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
import config
```

```python
# Connecting our postgresSQL database with our notebook
# Please make sure to update your postgres credential
engine = create_engine('postgresql://postgres:Jun3bug6@localhost:5432/caliaqi_db')

connection = engine.connect()
```

```python
# Creating Class and table
class Caqi(Base):
  __tablename__ = "breeze_aqi"
  city = Column(String, primary_key=True)
  aqi = Column(Integer)
  dominant_pol = Column(String)
  category = Column(String)
  pol_co = Column(Float)
  pol_o3 = Column(Float)
  pol_pm10 = Column(Float)
  pol_pm25 = Column(Float)
  pol_so2 = Column(Float)
  pol_no2 = Column(Float)
  time = Column(String)
```

```python
# Right now, this table only exist in python and not in the actual database
Base.metadata.tables
```

```python
# This is where we create our tables in the database
Base.metadata.create_all(engine)
```

```python
engine.table_names()
```

```python
# Push data to SQL
ca_air_quality_data.to_sql(name='breeze_aqi', con=engine, if_exists='replace', index=False)
```
